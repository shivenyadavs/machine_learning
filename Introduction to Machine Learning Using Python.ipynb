```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the iris dataset
iris_dataset = load_iris()

# Assuming x_train and y_train are your features and target labels
x_train = iris_dataset.data
y_train = iris_dataset.target

# Creating a DataFrame from the features
iris_dataframe = pd.DataFrame(x_train, columns=iris_dataset.feature_names)

# Plotting
pd.plotting.scatter_matrix(iris_dataframe, c=y_train, figsize=(15, 15), marker='o', hist_kwds={'bins': 20}, s=70, alpha=0.8, cmap='viridis')
plt.show()
```

# K-Nearest Neighbors

```python
from sklearn.neighbors import(KNeighborsClassifier)
knn=KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(x_train,y_train)
```

## **Making Predictions**

```python
x_new=np.array([[5,2.9,1,0.2]])
print("x_new.shape:{}".format(x_new.shape))
```

```python
prediction = knn.predict(x_new)
print("Prediction: {}".format(prediction))
print("Predicted target name: {}".format(iris_dataset['target_names'][prediction]))
```
